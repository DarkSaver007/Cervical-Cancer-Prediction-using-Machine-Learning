```python
import pandas as pd
import numpy as np
```

```python
cc=pd.read_csv("/content/drive/MyDrive/cancer_cervical (1).csv")
cc = cc.replace('?',np.nan)
```

```python
cc.rename(columns={'Number of sexual partners':'SP','Num of pregnancies':'Preg','Smokes (years)':'SMY','Hormonal Contraceptives':'HM','Hormonal Contraceptives (years)':'HMY','Smokes (packs/year)':'SMPY','IUD (years)':'IUDY','STDs (number)':'STDN',
                   'STDs:condylomatosis':'STD_CO','STDs:cervical condylomatosis':'STD_CE_CO','STDs:vaginal condylomatosis':'STD_VA_CO','STDs:vulvo-perineal condylomatosis':'STD_VP_CO','STDs:syphilis':'STD_S','STDs:pelvic inflammatory disease':'STD_PID',
                   'STDs:genital herpes':'STD_GH','STDs:molluscum contagiosum':'STD_MC','STDs:AIDS':'STD_AIDS','STDs:HIV':'STD_HIV','STDs:Hepatitis B':'STD_HEP','STDs:HPV':'STD_HPV','STDs: Number of diagnosis':'STD_NOD','STDs: Time since first diagnosis':'STD_TSFD',
                   'STDs: Time since last diagnosis':'STD_TSLD','Dx:Cancer':'DX_CAN','Dx:CIN':'DX_CIN','Dx:HPV':'DX_HPV','Dx':'DX','Hinselmann':'HIN','Schiller':'SCH','Citology':'CIT','Biopsy':'BIO','First sexual intercourse':'FSI'}, inplace=True)
```

```python
cc = cc.apply(lambda x: x.fillna(x.mode().iloc[0]))
cc
```

```python
cc  = cc.astype(float)
```

```python
X = cc.drop(columns = 'DX_CAN')
y= cc['DX_CAN']
```

```python
# Import necessary libraries
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score, precision_score, recall_score


# Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=45)

# Create an SVM classifier (you can experiment with different kernel types and parameters)
svm_model = SVC(kernel='linear', C=1.0)

# Train the model on the training set
svm_model.fit(X_train, y_train)

# Make predictions on the test set
y_pred = svm_model.predict(X_test)

# Evaluate the model
accuracy = accuracy_score(y_test, y_pred)
precision = precision_score(y_test, y_pred)
recall = recall_score(y_test, y_pred)

# Print the evaluation metrics
print(f"Accuracy: {accuracy:f}")
print(f"Precision: {precision:f}")
print(f"Recall: {recall:f}")
```
