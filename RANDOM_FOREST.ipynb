```python
import pandas as pd
import numpy as np
cc = pd.read_csv("/content/drive/MyDrive/cancer_cervical (1).csv")
cc =cc.replace('?',np.nan)
cc
```

```python
cc.rename(columns={'Number of sexual partners':'SP','Num of pregnancies':'Preg','Smokes (years)':'SMY','Hormonal Contraceptives':'HM','Hormonal Contraceptives (years)':'HMY','Smokes (packs/year)':'SMPY','IUD (years)':'IUDY','STDs (number)':'STDN',
                   'STDs:condylomatosis':'STD_CO','STDs:cervical condylomatosis':'STD_CE_CO','STDs:vaginal condylomatosis':'STD_VA_CO','STDs:vulvo-perineal condylomatosis':'STD_VP_CO','STDs:syphilis':'STD_S','STDs:pelvic inflammatory disease':'STD_PID',
                   'STDs:genital herpes':'STD_GH','STDs:molluscum contagiosum':'STD_MC','STDs:AIDS':'STD_AIDS','STDs:HIV':'STD_HIV','STDs:Hepatitis B':'STD_HEP','STDs:HPV':'STD_HPV','STDs: Number of diagnosis':'STD_NOD','STDs: Time since first diagnosis':'STD_TSFD',
                   'STDs: Time since last diagnosis':'STD_TSLD','Dx:Cancer':'DX_CAN','Dx:CIN':'DX_CIN','Dx:HPV':'DX_HPV','Dx':'DX','Hinselmann':'HIN','Schiller':'SCH','Citology':'CIT','Biopsy':'BIO','First sexual intercourse':'FSI'}, inplace=True)
```

```python
 cc =cc.apply(lambda x: x.fillna(x.mode().iloc[0]))
 cc
```

```python
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, classification_report

# Load the preprocessed DataFrame
df = pd.read_csv('/content/drive/MyDrive/cancer_cervical (1).csv')

target_column = 'DX_CAN'
X = cc.drop(columns=[target_column])
y = cc[target_column]

# Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=40)

# Initialize the Random Forest Classifier
rf_classifier = RandomForestClassifier(n_estimators=100, random_state=42)

# Train the classifier on the training data
rf_classifier.fit(X_train, y_train)

# Make predictions on the test set
y_pred = rf_classifier.predict(X_test)

# Evaluate the model
accuracy = accuracy_score(y_test, y_pred)
classification_rep = classification_report(y_test, y_pred)

print(f"Accuracy: {accuracy}")
print("Classification Report:\n", classification_rep)
```
