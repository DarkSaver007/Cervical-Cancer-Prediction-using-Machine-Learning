```python
import pandas as pd
```

```python
dt = pd.read_csv("/content/drive/MyDrive/cancer_cervical (1).csv")
import numpy as np
```

```python
dt = dt.replace('?',np.nan)
```

```python
 dt =dt.apply(lambda x: x.fillna(x.mode().iloc[0]))
 dt
```

```python
dt.rename(columns={'Number of sexual partners':'SP','Num of pregnancies':'Preg','Smokes (years)':'SMY','Hormonal Contraceptives':'HM','Hormonal Contraceptives (years)':'HMY','Smokes (packs/year)':'SMPY','IUD (years)':'IUDY','STDs (number)':'STDN',
                   'STDs:condylomatosis':'STD_CO','STDs:cervical condylomatosis':'STD_CE_CO','STDs:vaginal condylomatosis':'STD_VA_CO','STDs:vulvo-perineal condylomatosis':'STD_VP_CO','STDs:syphilis':'STD_S','STDs:pelvic inflammatory disease':'STD_PID',
                   'STDs:genital herpes':'STD_GH','STDs:molluscum contagiosum':'STD_MC','STDs:AIDS':'STD_AIDS','STDs:HIV':'STD_HIV','STDs:Hepatitis B':'STD_HEP','STDs:HPV':'STD_HPV','STDs: Number of diagnosis':'STD_NOD','STDs: Time since first diagnosis':'STD_TSFD',
                   'STDs: Time since last diagnosis':'STD_TSLD','Dx:Cancer':'DX_CAN','Dx:CIN':'DX_CIN','Dx:HPV':'DX_HPV','Dx':'DX','Hinselmann':'HIN','Schiller':'SCH','Citology':'CIT','Biopsy':'BIO','First sexual intercourse':'FSI'}, inplace=True)
```

```python
dt  = dt.astype(float)
```

```python
X = dt.drop(columns = 'DX_CAN')
Y= dt['DX_CAN']
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(X,Y,test_size=0.3,random_state=28)
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
```

```python
x_train = sc.fit_transform(x_train)
x_test= sc.fit_transform(x_test)
```

```python
from sklearn.tree import DecisionTreeClassifier
dt = DecisionTreeClassifier(criterion = "entropy")
dt.fit(x_train,y_train)
```

```python
dt_pred = dt.predict(x_test)
dt_pred
```

```python
from sklearn.metrics import accuracy_score,confusion_matrix
acc = accuracy_score(dt_pred,y_test)*100
```
